####**16. Saisie de chaînes**
+  Pour dire à R que quelque chose est une chaîne, vous l’entourez de guillemets doubles. 

+ Mais, que se passe-t-il si vous devez entrer une chaîne qui contient des guillemets doubles? 

+ Vous pouvez essayer d’utiliser des guillemets doubles, mais R renverra une erreur. 

+ R a interprété le deuxième guillemet double comme la fin de la chaîne, et est ensuite confus que d’autres entrées l’ont suivi.

####**17. Saisie de chaînes**
+ Nous pouvons éviter ce problème en utilisant des guillemets simples, ce que R permet également de définir une chaîne. R n’arrêtera pas d’interpréter l’entrée comme une chaîne jusqu’à ce qu’elle atteigne le guillemet simple de fermeture '. 

+ Examinez de près la sortie. R imprime toujours les chaînes avec des guillemets doubles (même si vous avez utilisé des guillemets simples pour les définir). 

+ Les guillemets doubles à l’intérieur de la chaîne ont une barre oblique inverse devant eux.

####**18. Saisie de chaînes**

+ Ceci est connu comme une séquence d’échappement, il indique à R qu’il s’agit d’un guillemet double, plutôt que c’est la fin de la chaîne. 

+ Échapper les guillemets doubles est une autre façon de les entrer dans une chaîne. Avec deux façons équivalentes d’entrer cette chaîne, laquelle devriez-vous utiliser?

####**19. Quand utiliser \ » vs. '**


+ R s’en fiche, la sortie est la même, mais il existe des directives conçues pour garder votre code lisible. 

+ Si le texte de la chaîne ne contient pas de guillemets, enveloppez-le entre guillemets. 

+ Si le texte contient des guillemets doubles mais pas des guillemets simples, enveloppez-le entre guillemets simples. 

+ **Si le texte contient les deux types de guillemets, enveloppez-le entre guillemets doubles et échappez les guillemets doubles dans le texte.**

###**Fractionnement des chaînes**


####**1. Fractionnement des chaînes**

+ **Lorsque vous utilisez str_detect, str_subset et str_count le modèle est d’intérêt central, vous voulez savoir où il se produit ou à quelle fréquence. Dans votre fonction de pigiste suivante, str_split, le motif n’est pas directement intéressant; Il fournit simplement un moyen utile de diviser la ficelle en morceaux.**

####**2. str_split()**

+ Prenez cette corde Tom & Jerry, nous ne sommes pas vraiment intéressés par

####**3. str_split()**


+ l’esperluette, mais il fournit un moyen utile de diviser la chaîne en deux caractères. 

+ Nous passons la chaîne et le motif, et la sortie est une liste avec un vecteur de chaînes, un élément pour chaque partie qui a été divisée, dans ce cas, un vecteur avec deux éléments, la chaîne « Tom » et la chaîne « Jerry ». 

+ str_split se divisera en autant de morceaux que possible, en fonction du nombre de fois que le motif se produit, nous nous retrouvons donc ici avec trois parties. 

+ Si vous souhaitez limiter le nombre de fractionnements, vous pouvez spécifier l’argument n. Si nous spécifions que n est égal à 2, la chaîne ne sera divisée qu’à la première occurrence du motif et nous récupérerons deux parties. 

####**4. str_split() renvoie une liste**

+ Vous vous demandez peut-être pourquoi str_split renvoie une liste. 

+ Eh bien, si nous passons un vecteur de chaînes, il n’y a aucune garantie que toutes les chaînes seront divisées en le même nombre de morceaux, une liste est la seule structure qui ne se plaindra pas si chaque élément est d’une longueur différente. 

+ Bien sûr, si vous spécifiez n ou si vous savez que toutes les chaînes ont le même nombre de morceaux, vous voudrez peut-être un type de sortie plus simple, comme une matrice.

####**5. str_split() peut renvoyer une matrice**

+  Vous pouvez demander ce type de sortie en spécifiant simplifier est égal à TRUE. 

+ Vous obtiendrez une matrice où chaque ligne correspond à une chaîne d’entrée et chaque colonne à un morceau de la division. 

+ Il y aura autant de colonnes que le plus grand nombre de fractionnements, et toutes les chaînes avec moins de fractionnements seront complétées par des chaînes vides. 

+ Spécifier à la fois n et simplifier garantit que vous n’aurez aucune surprise dans la dimension de la sortie: vous obtiendrez toujours n colonnes. Parfois, vous aurez besoin de la sortie de longueur variable.

####**6. Peignage avec lapply()**

+ **Pour traiter davantage la sortie non simplifiée, vous utiliserez généralement lapply ou sapply ou un équivalent. **

+ **Rappelez-vous que lapply prend une liste comme premier argument, et une fonction comme second.** 

+ **La fonction est appliquée à chaque élément de la liste et une liste est renvoyée.** 

+ **Par exemple, nous pourrions utiliser lapply pour trouver la longueur de chaque division, pour nous dire combien de morceaux chaque corde a été divisée.**

###**Références rétrospectives**

####**1. Références rétrospectives**

+ Maintenant que vous savez comment capturer des parties d’une expression régulière, vous êtes prêt à apprendre à faire référence aux éléments capturés. 

+ La référence à une partie capturée d’un motif est connue sous le nom de backreference.

####**2. Références rétrospectives**

+ Comme il peut y avoir plusieurs captures, les expressions régulières utilisent un simple comptage de gauche à droite pour faire référence aux différentes captures. 

+ **La première capture peut être référencée à l’aide de la constante REF1 dans rebus** : 
  + vous pouvez voir qu’elle se traduit par une barre oblique inverse 1 dans le langage d’expression régulière. 
  
+ **La deuxième capture peut être référencée en utilisant REF2 et ainsi de suite, jusqu’à la neuvième capture possible avec REF9.**

####**3. Dans un modèle**
+ Les références rétrospectives peuvent être utilisées directement dans un modèle, pour permettre des modèles répétés. 

+ Ce modèle recherche un mot en recherchant un espace, puis un ou plusieurs caractères de mot suivis d’un autre espace.

####**4. Dans un modèle**
+ Si nous capturons la partie du mot, nous pouvons nous y référer plus tard et le modèle reprend maintenant les mots répétés. 

####**5. En remplacement**
