####**1. Allocation de Dirichlet latente**

+ Nous avons vu comment le nombre de mots et les visualisations sugg√®rent quelque chose sur le contenu.

####**2. Apprentissage non supervis√©**


+ **Le mod√®le de sujet d‚Äôallocation latente de Dirichlet ou LDA recherche des mod√®les de mots apparaissant ensemble dans et √† travers un document de collection, √©galement connu sous le nom de corpus**. 

+ **Imaginez encore notre sac de mots. LDA trouve des sujets dans un corpus en cr√©ant un sac s√©par√© pour chaque document et en vidant les mots pour rechercher des mod√®les dans lesquels les mots apparaissent ensemble - pas seulement dans un sac, mais de mani√®re coh√©rente dans tous les sacs sp√©cifiques au document dans notre corpu**s. 

+ **Notez que LDA n‚Äôessaie pas d‚Äôexpliquer ou de pr√©dire une variable d√©pendante, comme dans une r√©gression**. 

+ **Au contraire, il s‚Äôagit simplement de rechercher des mod√®les au sein d‚Äôun groupe de variables explicatives**. 

+ **Cette recherche de mod√®le est connue sous le nom d‚Äôapprentissage non supervis√©.**

####**3. Probabilit√©s de mots**


+ **Les sujets eux-m√™mes sont une liste de tous les mots du corpus, souvent appel√©e dictionnaire, avec des probabilit√©s des mots apparaissant dans chaque sujet**. 

+ **Les mots qui apparaissent souvent ensemble auront de fortes probabilit√©s de se produire dans un ou plusieurs sujets.** 

+ Par exemple, nous avons ici deux sujets pour nos donn√©es d‚Äô√©valuation de produits.

####**4. Clustering vs mod√©lisation de sujets**

+ **Parce que la mod√©lisation de sujets est un type d‚Äôapprentissage non supervis√©, elle √©tablit naturellement des comparaisons avec d‚Äôautres techniques non supervis√©es**. 

+ **En particulier, la mod√©lisation th√©matique est souvent compar√©e au clustering, qui est courant dans la segmentation du march√© et une foule d‚Äôautres applications**. 

+ Il vaut la peine de prendre un moment pour distinguer ces deux techniques. 

+ **Les techniques de clustering courantes telles que les k-moyennes et le clustering hi√©rarchique sont bas√©es sur la distance entre les objets, qui est une mesure continue**. 

+ **En outre, chaque objet en cluster est affect√© √† un seul cluster**.

```r
# Run an LDA with 4 topics and a Gibbs sampler
lda_out3 <- LDA(
  dtm_twitter,
  k = 4,
  method = "Gibbs",
  control = list(seed = 42)
)
```

```r
lda_out3
```

```r
# Tidy the matrix of word probabilities
lda_topics3 <- lda_out3 %>% 
  tidy(matrix = "beta")
```

```r
head(lda_topics3)
```

```r
# Select the top 15 terms by topic and reorder term
word_probs3 <- lda_topics3 %>% 
  group_by(topic) %>% 
  top_n(15, beta) %>% 
  ungroup() %>%
  mutate(term2 = fct_reorder(term, beta))
```

```r
head(word_probs3)
```

```r
# Plot word_probs3, color and facet based on topic
ggplot(
  word_probs3, 
  aes(term2, beta, fill = as.factor(topic))
) +
  geom_col(show.legend = FALSE) +
  facet_wrap(~ topic, scales = "free") +
  coord_flip()
```
