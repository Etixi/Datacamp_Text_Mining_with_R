####**1. Tokenisation et nettoyage**

+ ***Vous pouvez compter les données catégorielles, mais le texte n’est toujours pas structuré. ####**2. Utilisation de tidytext**

+ ***C’est exactement ce que fait le paquet tidytext. Développé par Julia Silge et David Robinson, le package tidytext fournit une suite d’outils puissants qui nous permettent de structurer et d’analyser rapidement et facilement le texte, en tirant pleinement parti du tidyverse pour l’analyse de texte.***

####**3. Tokenisation du texte**

+ **Nous imposons une structure au texte en divisant chaque révision en mots distincts. **

+ **Dans le traitement du langage naturel ou les cercles NLP, cela s’appelle un sac de mots.**

+ **Nous ne nous soucions pas de la syntaxe ou de la structure des avis, nous coupons simplement chaque mot dans chaque avis et les mélangeons dans un sac: un sac de mots!** 

+ **Chaque corps de texte distinct est un document; Dans ce cas, les avis. Chaque mot unique est connu comme un terme. Toute occurrence d’un terme connu sous le nom de jeton; Ainsi, découper des documents en mots est connu sous le nom de tokenisation.**

####**4. Utilisation de unnest_tokens()**

+ **Après avoir chargé le paquet tidytext, la tokenisation est aussi simple que d’utiliser la fonction unnest_tokens()**. 

+ Après avoir spécifié le cadre de données d’entrée, nous fournissons le nom de la colonne de mots que nous créons par tokenisation, suivi du nom de la colonne avec le texte que nous voulons tokeniser. 

+ Dans review_data c’est la colonne de révision. 

+ Au lieu d’une colonne avec un examen dans chaque ligne, nous avons maintenant une colonne avec un seul mot dans chaque ligne. 

+ **En prime, unnest_tokens() a fait un peu de ménage pour nous: la ponctuation a disparu, chaque mot est en minuscules et les espaces blancs ont été supprimés.** 

+ **Le fait d’avoir un seul mot par ligne signifie que le nombre total de lignes dans le jeu de données est passé de 1 833 à 229 481.**

####**5. Compter les mots**

+ **Maintenant que nous avons imposé une structure ordonnée au texte, nous pouvons compter les mots en utilisant la fonction count().** 

+ **Pour faciliter la lecture des comptes, nous utilisons à nouveau le verbe arrange() et la fonction d’assistance desc()**. 

+ **Vous ne devriez pas être surpris de voir que les mots les plus couramment utilisés ne sont que des mots communs comme « le » qui ne donne pas beaucoup d’informations sur le contenu des critiques. Nous devons faire un nettoyage supplémentaire avant que notre nombre de mots ne soit informatif.**

####**6. Utilisation de anti_join()**

+ **Ces mots courants et non informatifs sont connus sous le nom de mots vides et nous aimerions les supprimer de notre cadre de données rangé**. 

+ **Un ensemble de fonctions en dplyr est pratique. Celles-ci sont connues sous le nom de jointures et, comme leur nom l’indique, elles sont utilisées pour joindre deux blocs de données ensemble en fonction d’une ou plusieurs colonnes correspondantes. La jointure que nous voulons s’appelle un anti_join(). Dans un anti_join(), une ligne du bloc de données de gauche est conservée tant que la valeur de la colonne correspondante n’est pas partagée par le bloc de données de droite.**

####**7. Utilisation de anti_join()**

+ Illustrons cela par review_data. 

+ **Un cadre de données de mots vides courants est disponible dans le package tidytext en tant que stop_words**. 

+ **Si nous canalisons le review_data tokenisé dans anti_join(), de sorte que les données de révision tokenisées soient le bloc de données gauche et stop_words soit le bon bloc de données, vous pouvez voir que le nombre de lignes a été considérablement _reduced._**

####**8. Recompter les mots**

+ Après avoir à nouveau calculé les comptes et les avoir organisés par ordre décroissant, nous pouvons maintenant voir que les mots les plus couramment utilisés dans les critiques de produits reflètent un contenu informatif réel.

###**Allocation de Dirichlet latente**


####**1. Allocation de Dirichlet latente**

+ Nous avons vu comment le nombre de mots et les visualisations suggèrent quelque chose sur le contenu. 

+ Nous avons également utilisé des dictionnaires de sentiments pour en savoir plus sur la valence émotionnelle d’un document. 

+ Dans ce dernier chapitre, nous allons au-delà du nombre de mots pour découvrir les sujets sous-jacents dans une collection de documents. Nous utiliserons un modèle de sujet standard connu sous le nom d’allocation de Dirichlet latente.

####**2. Apprentissage non supervisé**


+ **Le modèle de sujet d’allocation latente de Dirichlet ou LDA recherche des modèles de mots apparaissant ensemble dans et à travers un document de collection, également connu sous le nom de corpus**. 

+ **Imaginez encore notre sac de mots. LDA trouve des sujets dans un corpus en créant un sac séparé pour chaque document et en vidant les mots pour rechercher des modèles dans lesquels les mots apparaissent ensemble - pas seulement dans un sac, mais de manière cohérente dans tous les sacs spécifiques au document dans notre corpu**s. 

+ **Notez que LDA n’essaie pas d’expliquer ou de prédire une variable dépendante, comme dans une régression**. 

+ **Au contraire, il s’agit simplement de rechercher des modèles au sein d’un groupe de variables explicatives**. 

+ **Cette recherche de modèle est connue sous le nom d’apprentissage non supervisé.**

####**3. Probabilités de mots**


+ **Les sujets eux-mêmes sont une liste de tous les mots du corpus, souvent appelée dictionnaire, avec des probabilités des mots apparaissant dans chaque sujet**. 

+ **Les mots qui apparaissent souvent ensemble auront de fortes probabilités de se produire dans un ou plusieurs sujets.** 

+ Par exemple, nous avons ici deux sujets pour nos données d’évaluation de produits. 

+ **Le premier sujet est une collection de mots qui semblent être liés à la performance d’un aspirateur robotique, avec des mots comme cheveux, aspirateur, sols, maison et jour se produisant ensemble avec une forte probabilité.** 

+ **Le deuxième sujet est une collection de mots qui semblent être liés aux frustrations possibles d’un aspirateur robotique, avec des mots comme propre, nettoyage, bac et temps se produisant ensemble avec une forte probabilité.**

####**4. Clustering vs modélisation de sujets**

+ **Parce que la modélisation de sujets est un type d’apprentissage non supervisé, elle établit naturellement des comparaisons avec d’autres techniques non supervisées**. 

+ **En particulier, la modélisation thématique est souvent comparée au clustering, qui est courant dans la segmentation du marché et une foule d’autres applications**. 

+ Il vaut la peine de prendre un moment pour distinguer ces deux techniques. 

+ **Les techniques de clustering courantes telles que les k-moyennes et le clustering hiérarchique sont basées sur la distance entre les objets, qui est une mesure continue**. 

+ **En outre, chaque objet en cluster est affecté à un seul cluster**. 

+ **Les modèles thématiques comme LDA sont basés sur le nombre de mots, ce qui est une mesure discrète**. 

+ **De plus, chaque objet (dans ce cas, un document au sein d’un corpus) est un mélange ou un élément partiel de chaque sujet.**

###**Documenter les matrices terminologiques**
####**1. Documenter les matrices terminologiques**
+ Maintenant que nous avons une certaine intuition sur ce qu’est un modèle de sujet, et plus particulièrement sur ce qu’est LDA, préparons-nous à exécuter un modèle de sujet ! 

+ Pour ce faire, nous devrons nous familiariser avec la navigation dans et hors de notre cadre de données textuel bien rangé, en commençant par la création d’une matrice de termes de document.

####**2. Matrices et parcimonie**

+ Nous nous sommes concentrés sur les avantages de l’utilisation d’un cadre de données textuel ordonné, où chaque ligne est un seul mot ou jeton utilisé dans chaque document. 

+ Toutefois, pour exécuter un modèle de rubrique, nous devons d’abord créer une matrice de termes de document ou DTM. Pour illustrer, sparse_review est un sous-ensemble du DTM que nous allons créer. 

+ Il comprend le nombre unique de mots ou de termes pour seulement quatre avis et uniquement les termes commençant par A. 

+ Nous pouvons voir qu’une matrice est comme un cadre de données, sauf que chaque colonne doit être du même type. Nous indexons une matrice en nous référant à ses lignes puis à ses colonnes. 

+ Une matrice terminologique de document comporte une seule ligne pour chaque document et une colonne pour chaque mot ou terme unique utilisé dans tous les documents du corpus. 

+ Les valeurs dans le DTM sont le nombre de jetons ou d’utilisations de chaque terme pour le document donné. En regardant sparse_review, nous pouvons voir que la valeur la plus courante est zéro. 

+ Lorsque vous avez une matrice composée principalement de zéros, on parle de parcimonie ou de matrice clairsemée. 

+ Notre DTM sera probablement clairsemé, car la plupart des documents n’utilisent pas la plupart des termes présents dans le corpus.

####**3. Utilisation de cast_dtm()**

+ **Pour créer un DTM, prenons nos données tidy_review et comptons chaque mot dans chaque avis, indiqué ici par la colonne id**. 

+ **Nous pouvons maintenant utiliser la fonction cast_dtm() du paquet tidytext pour convertir facilement notre trame de données ordonnée en DTM.** 

+ **Pour utiliser cast_dtm(), nous devons spécifier la colonne du document, la colonne du terme et le nombre de mots, dans cet ordre**. 

+ Cependant, la sortie ici n’est pas très informative. 

+ On nous dit que le résultat est en effet un DTM, où il y a 1 791 documents ou revues et 9 669 termes, ou mots uniques. 

+ Il y a aussi une référence à la rareté, ou combien d’entrées dans cette matrice sont non nulles. Comme on pouvait s’y attendre, le DTM est très clairsemé.

####**2. Utilisation de LDA()**


+ Pour exécuter un modèle de rubrique, nous utiliserons le package topicmodels conjointement avec tidyverse et tidytext. 

+ **Avec un DTM comme entrée, l’exécution d’un modèle de rubrique est simple : nous utilisons la fonction LDA()**. 

+ Il y a quatre arguments pour cette fonction que nous devrons spécifier. 

  + **Le premier argument est l’entrée DTM**. 

  + **Le deuxième argument est k, ou le nombre de sujets que nous voulons que le modèle produise. Nous verrons plus tard comment nous pouvons décider de ce que nous devrions définir k.**

  + **Le troisième argument est la méthode d’estimation. La valeur par défaut est une approximation rapide. Cependant, si nous préférons une méthode plus longue mais plus complète, nous devrions spécifier l’échantillonneur de Gibbs**
  
  + **Enfin, le quatrième argument nous permet de spécifier la graine de simulation, que nous devons traiter comme une liste, une structure de données en R qui dépasse le cadre de ce cours**. 
  
  + **La spécification de la graine de simulation nous aidera à récupérer des sujets cohérents sur des modèles répétés, compte tenu de la nature probabiliste de l’estimation du modèle.**

####**3. Sortie LDA()**

+ Après avoir exécuté le modèle, ce qui peut prendre quelques instants ou quelques heures, selon la quantité de texte que nous lui lançons, nous pouvons voir que la sortie est, tout comme le DTM, un objet R codé spécifiquement pour le package topicmodels.

####**5. Utilisation de tidy()**

+ Nous aimerions évaluer la sortie du modèle d’une manière cohérente avec notre suite d’outils ordonnés. 

+ En particulier, les résultats les plus importants d’un modèle de sujet sont les sujets eux-mêmes: 
  + le dictionnaire de tous les mots du corpus triés en fonction de la probabilité que chaque mot apparaisse dans le cadre de ce sujet. 
  + Encore une fois, nous pouvons utiliser une fonction tidytext, dans ce cas tidy(), pour prendre la matrice des probabilités de sujet et les mettre sous une forme facilement visualisable à l’aide de ggplot2. 
    + L’utilisation de tidy() nous oblige à spécifier le nom et la structure du morceau de lda_out que nous voulons ranger; Dans ce cas, il s’agit d’une « matrice » et d’une « bêta ». 
    + Si cast_dtm() nous permet de naviguer hors des formats de données ordonnés pour exécuter le modèle de rubrique, tidy() nous permet de prendre la sortie du modèle et de revenir à un format de données ordonné.

###**Interprétation des sujets**

####**1. Interprétation des sujets**


+ ’interprétation d’un modèle thématique est une forme d’art. 

+ Tout comme le regroupement et d’autres techniques d’apprentissage non supervisé, nous obtenons une description de ce qui compose chaque sujet, mais aucune autre direction quant à la signification des sujets. 

+ De même, bien que nous puissions utiliser une variété de diagnostics pour essayer de décider du nombre de sujets à estimer, la décision revient en fin de compte à celui qui effectue l’analyse. 

+ La clé est de trouver des sujets qui sont tous différents où les sujets ne se répètent pas.

####**2. Deux thèmes**
+ Ici, nous exécutons LDA avec k = 2 sujets et canalisons la sortie du modèle directement dans tidy() pour extraire le mot probabilités. 

+ Nous sélectionnons ensuite les mots les plus probabilités et réorganisons les termes par sujet pour produire word_probs.

####**3. Deux thèmes**
+ Enfin, nous traçons word_probs, en traitant le sujet comme un facteur pour ajouter de la couleur, en supprimant la légende redondante de l’intrigue à barres et en les facettant par sujet pour voir facilement ce qui rend les sujets différents. 

+ Pour interpréter ces sujets, nous devons considérer ce que les mots apparaissant avec une forte probabilité dans chaque sujet suggèrent. 

+ Encore une fois, il s’agit d’une décision subjective. 

+ Le premier sujet est une collection de mots qui semblent être liés à la performance d’un aspirateur robotique, avec des mots comme cheveux, aspirateur, sols, maison et jour se produisant ensemble avec une forte probabilité. 

+ Le deuxième sujet est une collection de mots qui semblent être liés à des frustrations possibles avec la fonctionnalité d’un aspirateur robotique, avec des mots comme propre, nettoyage, bac et temps se produisant ensemble avec une forte probabilité. 

+ Nous pourrions nommer le premier sujet « Performance » et le deuxième sujet « Fonctionnalité ».
