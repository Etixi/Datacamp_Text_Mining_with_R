####**4. Analyse sémantique vs sac de mots**

+ Découvrons maintenant deux approches de l’exploration de texte. 
+ **Le premier est l’analyse sémantique basée sur la syntaxe des mots**. 
  + **Dans l’analyse sémantique, vous vous souciez du type et de l’ordre des mots. Cette méthode crée beaucoup de fonctionnalités à étudier.** 
    + Par exemple, un seul mot peut être étiqueté dans le cadre d’une phrase, puis un nom et aussi un nom propre ou une entité nommée. 
    
+ Donc, ce seul mot a trois caractéristiques qui lui sont associées. Cet effet rend l’analyse sémantique « riche en fonctionnalités ». 

+ Pour effectuer le balisage, l’analyse sémantique suit une arborescence pour diviser continuellement le texte. 

+ En revanche, la méthode du sac de mots ne se soucie pas du type ou de l’ordre des mots. 

+ Ici, les mots ne sont que des attributs du document. Dans cet exemple, nous analysons la phrase « Steph Curry a raté un coup dur ». 

+ Dans l’exemple sémantique, vous voyez comment les mots sont décomposés de la phrase, aux phrases de nom et de verbe et finalement en attributs uniques. 

+ **Bag of words traite chaque terme comme un seul jeton dans la phrase, quel que soit le type ou l’ordre**.

####**Corpus**
Pour faire un corpus, les données (vecteur ou data frame) doivent d'abord être interprétées comme un document :

  + **VectorSource(): interprète chaque élément du vecteur $x$ comme un document.**

  + **DataframeSource(): interprète chaque ligne du bloc de données $x$ comme un document**. 
  
    + La première colonne doit être nommée "doc_id" et contenir un identifiant de chaîne unique pour chaque document. 
    
    + La deuxième colonne doit être nommée "texte" et contenir une chaîne encodée en UTF-8 représentant le contenu du document. Des colonnes supplémentaires facultatives sont utilisées comme métadonnées au niveau du document.

+ Il existe deux types de données de corpus, **le corpus permanent, PCorpus, et le corpus volatile, VCorpus.**


+ **Essentiellement, la différence entre les deux a à voir avec la façon dont la collection de documents est stockée sur votre ordinateur.**


+ **Nous utiliserons le corpus volatil, qui est conservé dans la RAM de votre ordinateur plutôt que enregistré sur le disque, juste pour être plus efficace en mémoire.
VCorpus crée des corpus volatils**

```r
# two lines:
amzn_pros_source <- VectorSource(amzn_pros)
amzn_pros_corpus <- VCorpus(amzn_pros_source)

# or as one liners:
amzn_cons_corpus <- VCorpus(VectorSource(amzn_cons))
goog_pros_corpus <- VCorpus(VectorSource(goog_pros))
goog_cons_corpus <- VCorpus(VectorSource(goog_cons))
```

[Sac de mots/Bag of Words](#https://lbourdois.github.io/blog/nlp/Bag-of-word/)

  + **Sac de mots : regroupez tous les mots du vecteur en une seule chaîne, puis combinez-les**

```r
# each vector contains two "bags": pros and cons
all_amzn <- c(paste(amzn_pros, collapse = " "), paste(amzn_cons, collapse = " "))
names(all_amzn) <- c("amzn_pros", "amzn_cons")

all_goog <- c(paste(goog_pros, collapse = " "), paste(goog_cons, collapse = " "))
names(all_goog) <- c("goog_pros", "goog_cons")

# bags can be conbined together
reviews <- c(all_amzn, all_goog)

# make corpus
reviews_corpus <- VCorpus(VectorSource(reviews))
inspect(reviews_corpus)
```

```r
# Find the 10 most frequent terms: term_count
term_count <- freq_terms(reviews_corpus, 10)
plot(term_count)
```

####**Introduction à la radicalisation des mots et à l'achèvement de la radicalisation**

+ ***Pourtant, une autre étape de prétraitement utile implique la radicalisation des mots et l'achèvement de la radicalisation. La radicalisation des mots réduit les mots pour les unifier entre les documents. Par exemple, la racine de "computational", "computers" et "computation" est "comput". Mais parce que "comput" n'est pas un vrai mot, nous voulons reconstruire les mots de sorte que "computational", "computers" et "computation" se réfèrent tous à un mot reconnaissable, tel que "computer". L'étape de reconstruction s'appelle l'achèvement de la tige.***

+ **Le package tm fournit la fonction stemDocument() pour accéder à la racine d'un mot. Cette fonction prend soit un vecteur de caractères et renvoie un vecteur de caractères, soit prend un PlainTextDocument et renvoie un PlainTextDocument.**

+ Par exemple,

      stemDocument(c("computational", "computers", "computation"))
      returns "comput" "comput" "comput".

+ *Vous utiliserez stemCompletion() pour reconstruire ces racines de mots en un terme connu. stemCompletion() accepte un vecteur de caractères et un dictionnaire de complétion. Le dictionnaire de complétion peut être un vecteur de caractères ou un objet Corpus. Quoi qu'il en soit, le dictionnaire de complétion de notre exemple devrait contenir le mot "computer", afin que toutes les instances de "comput" puissent être reconstruites.*

####**Modification des pondérations de fréquence**


+ ***Jusqu'à présent, vous avez simplement compté les termes dans les documents dans DocumentTermMatrix ou TermDocumentMatrix. Dans cet exercice, vous découvrirez la pondération TfIdf au lieu de la simple fréquence des termes. TfIdf signifie fréquence de terme-fréquence de document inverse et est utilisé lorsque vous avez un grand corpus avec une diversité de terme limitée.***

+ **TfIdf compte les termes (c'est-à-dire Tf), normalise la valeur en fonction de la longueur du document, puis pénalise la valeur plus un mot apparaît souvent parmi les documents. C'est du bon sens; si un mot est banal, il est important mais pas perspicace. Cet aspect de pénalité est capturé dans la fréquence de document inverse (c'est-à-dire, Idf).**

+ *Par exemple, l'examen des notes du service client peut inclure le terme "cu" comme raccourci pour "client". Une note peut indiquer "le cu a un colis endommagé" et une autre comme "cu a appelé avec une question sur la livraison". Avec la pondération fréquentielle du document, "cu" apparaît deux fois, il est donc censé être informatif. Cependant, dans TfIdf, « cu » est pénalisé car il apparaît dans tous les documents. Par conséquent, "cu" n'est pas considéré comme nouveau, sa valeur est donc réduite vers 0, ce qui permet aux autres termes d'avoir des valeurs plus élevées pour l'analyse.*

```r
# Clean with qdap
qdap_clean <- function(x) {
  x <- replace_abbreviation(x)
  x <- replace_contraction(x)
  x <- replace_number(x)
  x <- replace_ordinal(x)
  x <- replace_symbol(x)
  x <- tolower(x)
  return(x)
}

# Clean with tm
tm_clean <- function(corpus) {
  corpus <- tm_map(corpus, removePunctuation)
  corpus <- tm_map(corpus, stripWhitespace)
  corpus <- tm_map(corpus, removeWords, 
                   c(stopwords("en"), "Google", "Amazon", "company"))
  return(corpus)
}
```

```r
qdap_clean <- function(x){
  x <- replace_abbreviation(x)
  x <- replace_contraction(x)
  x <- replace_number(x)
  x <- replace_ordinal(x)
  x <- replace_ordinal(x)
  x <- replace_symbol(x)
  x <- tolower(x)
  return(x)
}
```

```r
clean_text <- function(corpus){
  corpus <- tm_map(corpus, content_transformer(qdap_clean))
  corpus <- tm_map(corpus, removePunctuation)
  corpus <- tm_map(corpus, stripWhitespace)
  corpus <- tm_map(corpus, removeWords, c(stopwords("en"), "google", "amazon", "company"))
  return(corpus)
}

reviews_clean <- clean_text(reviews_corpus)
```

```r
# Create a term-document matrix from the corpus
reviews_tdm <- TermDocumentMatrix(reviews_clean)

# Print reviews_tdm data
reviews_tdm
```

```r
top_df <- reviews_m[,c("amzn_cons", "goog_cons")] %>%
  # Convert to data frame
  as_data_frame(rownames = "word") %>% 
  # Keep rows where word appears everywhere
  filter_all(all_vars(. > 0)) %>% 
  # Get difference in counts
  mutate(difference = amzn_cons - goog_cons) %>% 
  # Keep rows with biggest difference
  top_n(5, wt = difference) %>% 
  # Arrange by descending difference
  arrange(desc(difference))

# library(plotrix) is loaded
pyramid.plot(
  # Chardonnay counts
  top_df$amzn_cons, 
  # Coffee counts
  top_df$goog_cons, 
  # Words
  labels = top_df$word, 
  top.labels = c("Amazon", "Words", "Google"), 
  main = "Words in Common", 
  unit = NULL,
  gap = 40
)
```
